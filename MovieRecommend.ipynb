```python
def convertCast(obj):
    if obj == np.nan or obj == "Random":
        return list(("Random",))
    else:
        l = []
        counter = 0
        for i in ast.literal_eval(obj):
            if counter != 3:
                l.append(i["name"])
                counter += 1
            else:
                break
        return l
```

```python
movies_df["cast"]=movies_df["cast"].apply(convertCast)
```

```python
movies_df
```

```python
def fetchDirector(obj):
    if obj == np.nan or obj == "Random":
        return list(("Random",))
    else:
        l = []
        counter = 0
        for i in ast.literal_eval(obj):
           if i["job"]== "Director":
               l.append(i["name"])
               break
        return l
```

```python
movies_df["Director"] = movies_df["crew"].apply(fetchDirector)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(movies_df_final["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = movies_df_final[movies_df_final["title"] == movie].index
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances[0])), reverse=True, key=lambda x:x[1])[1:6]
    for i in movie_list:
        # print(movies_df_final.iloc[i[0]]["title"])
        print(i)
        print(movies_df_final.iloc[i[0]]["id"])
```
